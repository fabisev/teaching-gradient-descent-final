# Gradient Descent Demo Notebook

This is the notebook-native version of the gradient descent demo. It is intended for use in Jupyter rather than as the primary built-book interactive.

Use the sliders to explore how the learning rate, starting point, and number of steps affect convergence on a simple one-dimensional loss.

```python
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as widgets
from IPython.display import display

plt.style.use('seaborn-v0_8-whitegrid')
```

```python
def loss(x):
    return x ** 2

def gradient(x):
    return 2 * x

def gradient_descent(start, learning_rate, steps):
    xs = [start]
    for _ in range(steps):
        next_x = xs[-1] - learning_rate * gradient(xs[-1])
        xs.append(next_x)
    return np.array(xs)

def plot_descent(learning_rate=0.2, start=4.0, steps=10):
    xs = gradient_descent(start, learning_rate, steps)
    curve_x = np.linspace(-5, 5, 400)
    curve_y = loss(curve_x)

    fig, ax = plt.subplots(figsize=(8, 5))
    ax.plot(curve_x, curve_y, label='f(x) = x^2', color='steelblue')
    ax.scatter(xs, loss(xs), color='crimson', zorder=3, label='Gradient descent steps')

    for i in range(len(xs) - 1):
        ax.annotate(
            '',
            xy=(xs[i + 1], loss(xs[i + 1])),
            xytext=(xs[i], loss(xs[i])),
            arrowprops={'arrowstyle': '->', 'color': 'crimson', 'lw': 1.5}
        )

    ax.set_title('Gradient descent on f(x) = x^2')
    ax.set_xlabel('x')
    ax.set_ylabel('f(x)')
    ax.legend()
    plt.show()

    print(f'Final x: {xs[-1]:.4f}')
    print(f'Final loss: {loss(xs[-1]):.6f}')

controls = widgets.interactive(
    plot_descent,
    learning_rate=widgets.FloatSlider(value=0.2, min=0.01, max=1.2, step=0.01, description='LR'),
    start=widgets.FloatSlider(value=4.0, min=-5.0, max=5.0, step=0.1, description='Start'),
    steps=widgets.IntSlider(value=10, min=1, max=30, step=1, description='Steps')
)

display(controls)
```
